**TEXTBOX &** **BUTTON**

```python
text_box =pn.widgets.TextInput(name='Enter name:- ' ,value='')
text =pn.widgets.StaticText()

#creating function
def hello(event):
    text.value = f'Hello {text_box.value}. Have a great day!'
    return text

# create button widget
button = pn.widgets.Button(name='Click ME', button_type='success')
button.clicks

# button with on click
button.on_click(hello)

# layout
pn.Column(text_box, button, text)
```

```python
from vega_datasets import data as vds
```

```python
stocks = vds.stocks()
stocks.head()
```

```python
# symbols
symbols = list(stocks.symbol.unique())
symbols
```

```python
select= pn.widgets.Select(name='select',options=symbols)

# creating function
def create_plot(symbol):
    return stocks[stocks['symbol'] == symbol].hvplot('date', 'price')

# create interaction between widget and function
interact(create_plot, symbol=select)
```

**CHECKBOX**

```python
# create widget
checkbox = pn.widgets.Checkbox(name='Checkbox')

# create function
@pn.depends(checkbox)#using depends decorator
def checked(x):
    if x == True:
        return 'checked'
    else:
        return 'not checked'

# create interaction between widget and function
pn.Column(pn.Column(checkbox), checked)
```
